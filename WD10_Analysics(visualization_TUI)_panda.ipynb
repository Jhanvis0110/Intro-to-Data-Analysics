```python
def cGST(arg):
  if arg<=10:
    pass
  elif arg<=20:
    arg=arg*105/100
  elif arg<=30:
    arg=arg*110/100
  elif arg>30:
    arg=arg*120/100
  else:
    arg="invalid data"
  return arg
```

```python
df["New_GST"]=df.Total.apply(cGST)
```
